**1. Import Matplotlib & Numpy**


```
#Your code here
import numpy as np
import matplotlib.pyplot as plt
```

**2. Preparing data**

Create x vector which has 100 elements and the element values gradually increase from 0 to 99.

Randomly create y vector which has 100 elements and the element values are randomly generated in the range of \[0, 1000)

```
#Your code here
np.random.seed(1)
x = np.arange(0, 100, 1)
y = np.random.randint(0, 1000, 100)
print(len(x))
print(len(y))
```

**3. Draw a line chart**

Use the above data to draw a line chart.

```
#Your code here
plt.plot(x, y)
plt.show()
```

**4. Config the axes**

Use the above generated dataset, let set the name for x-axis, y-axis and the legend of your chart.
Suggested reading: [link](https://www.tutorialspoint.com/matplotlib/matplotlib_axes_class.htm)

```
#Your code 
plt.plot(x, y, label='line 1')
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("My first line graph")
plt.legend()
plt.show()
```

**5. Multiple line chart**

Create x vector which has 100 elements and the element values gradually increase from 0 to 99.

Randomly create y1 and y2 vector which have 100 elements and the element values are randomly generated in the range of \[0, 1000) 

Let draw the above two lines y1 and y2. The chart must has:


1.   names of x-axis and y-axis
2.   chart legend
3.   different colors for the lines of y1 and y2
4.   different markers for the lines of y1 and y2





```
#Your code here
x = np.arange(0, 100, 1)
y1 = np.random.randint(0, 1000, 100)
y2 = np.random.randint(0, 1000, 100)

plt.plot(x, y1, 'ys-', label='line 1' )
plt.plot(x, y2, 'go--', label='line 2')
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("My second line graph")
plt.legend()
plt.show()
```

**6. Draw a box plot**

Use the above y1, y2 vectors to draw a box plot.

```
#Your code here
data_to_box_plot = [y1, y2]

plt.boxplot(data_to_box_plot)
plt.show()
```

**7. Draw a histogram**

Suggested reading: [link](https://www.tutorialspoint.com/matplotlib/matplotlib_histogram.htm). You make your own data.

```
#Your code here
y = 5 + np.random.randn(1000)

plt.hist(y, bins=20)
plt.title("Histogram")
plt.show()

plt.hist(y, cumulative=True, bins=20)
plt.title("Cumulative Histogram")
plt.show()
```

**8. Draw a bar plot**

Suggested reading: [link](https://www.tutorialspoint.com/matplotlib/matplotlib_bar_plot.htm). You make your own data.

```
#Your code here
plt.bar(x, y1, label="Blue Bar", color='b')
plt.bar(x, y2, label="Green Bar", color='g')
plt.plot()

plt.xlabel("bar number")
plt.ylabel("bar height")
plt.title("Bar Chart Example")
plt.legend()
plt.show()
```

**9. Draw a scatter plot**

Use the above x, y1, y2 vectors to draw a scatter plot. Suggested reading: [link](https://www.tutorialspoint.com/matplotlib/matplotlib_scatter_plot.htm)

```
#Your code here
plt.scatter(x, y1, marker='v', color='r')
plt.scatter(x, y2, marker='^', color='m')
plt.title('Scatter Plot Example')
plt.show()
```

**10. Subplot**

Combine the plots of sections 4, 5, 6, 7, 8, and 9 into a multi-plot. Suggested reading: [link](https://www.tutorialspoint.com/matplotlib/matplotlib_subplots_function.htm)

```
#Your code here
fig, a =  plt.subplots(3,2)

a[0][0].plot(x, y1, label='line 1')
a[0][0].set_xlabel("x axis")
a[0][0].set_ylabel("y axis")
a[0][0].set_title("My first line graph")
a[0][0].legend()

a[0][1].plot(x, y1, 'ys-', label='line 1' )
a[0][1].plot(x, y2, 'go--', label='line 2')
a[0][1].set_xlabel("x axis")
a[0][1].set_ylabel("y axis")
a[0][1].set_title("My second line graph")
a[0][1].legend()

a[1][0].boxplot(data_to_box_plot)

a[1][1].hist(y, cumulative=True, bins=20)
a[1][1].set_title("Cumulative Histogram")

a[2][0].bar(x, y1, label="Blue Bar", color='b')
a[2][0].bar(x, y2, label="Green Bar", color='g')
a[2][0].set_xlabel("bar number")
a[2][0].set_ylabel("bar height")
a[2][0].set_title("Bar Chart Example")
a[2][0].legend()

a[2][1].scatter(x, y1, marker='v', color='r')
a[2][1].scatter(x, y2, marker='^', color='m')
a[2][1].set_title('Scatter Plot Example')

plt.tight_layout()
plt.show()

```
